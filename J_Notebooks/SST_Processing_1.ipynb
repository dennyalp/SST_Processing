```python
# Import required libraries 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import os 
```

```python
filePath = "C:/Users/denny/Science/Data_science/SSTs_Processing/SST_Processing/data/CASPER_ssst_clean.txt"
dfSST = pd.read_table(filePath, sep = "\s+", header = 0, 
                      names = ["nrec", "yr", "mm", "dd", "HH", "MM", "SS", "Lat", "Long", "Ts"],
                      skiprows=1)
```
